```python
import pandas as pd
import numpy as np
#!pip install selenium
#!pip install BeautifulSoup4
from bs4 import BeautifulSoup  # to scrap web
```

```python
import ssl  #to certify links
ssl._create_default_https_context = ssl._create_unverified_context 
```

```python
 import requests
 webpage=requests.get('https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports').text

 soup=BeautifulSoup(webpage,'lxml') #to convert it in readable form [soupify process]
 date=[]
 for i in soup.find_all('a',class_="js-navigation-open Link--primary"):
   #print(i.text[0:10].strip())
   date.append(i.text[0:10].strip())
    
 def scrapper(date):
     url="https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_daily_reports/"+date
     ii=".csv"
     furl=url+ii+"?raw=true"  #raw(arranged) ko thik karna  (concatenate)
     fetch = pd.read_csv(furl)
     return fetch

 for i in date[1:len(date)-1]:
     df=scrapper(i)
     df.to_csv("covidall.csv", mode='a', index=False, header=df.columns)  # to convert df to csv file
     #header fun is used to get the same col as in df 
     print(f"{i} --> Data Fetched")
```
